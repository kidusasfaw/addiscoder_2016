```python
from sklearn import tree
import numpy as np
```

### Random Forest Regressor (because it pivots on each variable)

```python
# get our input data
column_names = ['Q' + str(n) for n in range(1,16) if n != 12]
X = df[column_names]
X = np.array(X)

# correct NAN's
for row_i in range(len(X)):
    for elem_i in range(len(X[row_i])):
        if X[row_i][elem_i] != 1:
            X[row_i][elem_i] = 0

# our output data
y = np.array(df['Total'])

# our tree regressor
clf = tree.DecisionTreeRegressor()
clf = clf.fit(X, y)

# sanity checks
print(clf.predict([[0 for i in range(1,15)]]))
print(clf.predict([[1 for i in range(1,15)]]))

# what we came for
importances = sorted([('Q' + str(i + 1), clf.feature_importances_[i]) for i in range(len(clf.feature_importances_))],
                     key=(lambda x : -x[1]))
for (q, importance_score) in importances:
    print(q + ' had a score of ' + str(importance_score))
```

### Do we get something different by using a classifier instead?

```python
# get our input data
column_names = ['Q' + str(n) for n in range(1,16) if n != 12]
X = df[column_names]
X = np.array(X)

# correct NAN's
for row_i in range(len(X)):
    for elem_i in range(len(X[row_i])):
        if X[row_i][elem_i] != 1:
            X[row_i][elem_i] = 0

# our output data
y = np.array([str(x) for x in df['Total']])

# our tree regressor
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)

# sanity checks
print(clf.predict([[0 for i in range(1,15)]]))
print(clf.predict([[1 for i in range(1,15)]]))

# what we came for
importances = sorted([('Q' + str(i+1), clf.feature_importances_[i]) for i in range(len(clf.feature_importances_))],
                     key=(lambda x : -x[1]))
for (q, importance_score) in importances:
    print(q + ' had a score of ' + str(importance_score))
```

## Or we can try a bunch of linear regressions to see whether a variable is good at predicting the score

```python
from sklearn import linear_model

questions = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8', 'Q9', 'Q10', 'Q11', 'Q13', 'Q14']
fit_data = []
intercepts = []
for question_number in questions:
    # Clean up data into the right format
    X = np.array([[x] for x in df[question_number]])
    for i in range(len(X)): #NaN struggles
        if X[i] != 1:
            X[i] = 0
    y = np.array(df['Total'])

    reg = linear_model.LinearRegression()
    reg.fit(X,y)
    print('Slope is ' + str(reg.coef_[0]))
    print('Intercept is ' + str(reg.intercept_))
    print('R-squared is ' + str(reg.score(X,y)))

    plt.scatter(X,y)
    plt.plot(X, reg.predict(X), color='blue', linewidth=3)
    plt.xticks(())
    plt.yticks(())
    plt.title(str(question_number))
    plt.show()
    fit_data.append({ 'name': question_number, 'slope': reg.coef_[0], 'r2': reg.score(X,y) })
    intercepts.append(reg.intercept_)

print('Background score is ' + str(np.mean(intercepts)))
sorted(fit_data, key=(lambda x: -x['r2']))
```

We see that higher $r^2$ values correspond to higher slopes as well

## What if we try PCA on this data?

```python
from sklearn import decomposition

# Clean data
column_names = ['Q' + str(n) for n in range(1,16) if n != 12]
X = df[column_names]
X = np.array(X)
# correct NAN's
for row_i in range(len(X)):
    for elem_i in range(len(X[row_i])):
        if X[row_i][elem_i] != 1:
            X[row_i][elem_i] = 0

# 2-D decomp and divide into three buckets
pca = decomposition.PCA(n_components=2)
pca.fit(X)
print(pca.explained_variance_)
x_plot = [[],[],[]]
y_plot = [[],[],[]]
x2d = pca.fit_transform(X)
for i in range(len(x2d)):
    if y[i] < 6:
        x_plot[0].append(x2d[i][0])
        y_plot[0].append(x2d[i][1])
    elif y[i] < 9:
        x_plot[1].append(x2d[i][0])
        y_plot[1].append(x2d[i][1])
    else:
        x_plot[2].append(x2d[i][0])
        y_plot[2].append(x2d[i][1])
    
plt.scatter(x_plot[0],y_plot[0],color='green',label='scores 0 through 5')
plt.scatter(x_plot[1],y_plot[1],color='blue',label='scores 6 through 8')
plt.scatter(x_plot[2],y_plot[2],color='red',label='scores 9+')
plt.legend()
plt.title('2-D PCA decomp')
plt.show()

# 1-D decomp with regression
pca1 = decomposition.PCA(n_components=1)
pca1.fit(X)
print(pca1.explained_variance_)
x1d = pca1.fit_transform(X)
reg = linear_model.LinearRegression()
reg.fit(x1d,y)
plt.scatter(x1d, y)
plt.plot(x1d,reg.predict(x1d),color='blue',linewidth=2)
plt.title('1-D decomp vs. score')
plt.show()
print('R-squared ' + str(reg.score(x1d,y)))
```

So basically 1-d PCA preserves most of the information
