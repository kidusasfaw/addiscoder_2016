### Problem 4
Write a function `isSorted` which takes as input a list `L` of integers and returns `True` if `L` is sorted and `False` if it is not sorted.

````python
def isSorted(l):
    for i in range(len(l)-1):
        if l[i]<l[i+1]:
            #return True
            continue
        else:
            return False
    return True
isSorted([1,2,3,4,5])
````
